```python
#to filter rows of a panda data frame by column value
import pandas as pd
movies =pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
booleans =[]
for length in movies.duration:
    if length >= 200:
        booleans.append('True')
    else:
        
        booleans.append('False')
```

```python
booleans[0:7]
```

```python
#simpler way to get the resultant output for filtering the series
movies[movies.duration >= 200]
```

```python
#to pull out a specific series from a data frame from the filtering
movies[movies.duration >= 200].genre #or movies[movies.duration >= 200]['genre']
```

```python
# to use loc for more accurate or the best way is to use loc
movies.loc[movies.duration >= 200,'genre'] #loc is used to fetch correct output by giving the rows and colums we want in the result
```

```python
#to use multiple filter criteria in a data frame using pandas
import pandas as pd
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
movies[movies.genre == 'Crime']
```

```python
movies[movies.duration >= 200 & movies.genre == 'Crime'] #wrog syntax should use paranthesis 
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Crime')]
```

```python
#using or operator
movies[(movies.duration >= 200 ) |(movies.star_rating > 9)]
```

```python
#to select rows having multiple or or and conditions
movies[(movies.star_rating >= 9 ) & (movies.genre in ['Drama','Action','Crime'])] #wrong syntax
```

```python
movies[(movies.duration >= 200) & (movies.genre.isin(['Drama','Action','Crime']))]
```
